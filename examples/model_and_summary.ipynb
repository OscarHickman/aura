# Model and Summary

This notebook shows core recommendation and summary features from Python code.

```python
import os
import sys
from pathlib import Path
from pprint import pprint

# Ensure the repository root is importable when running notebook from examples/.
repo_root = Path.cwd().resolve().parent
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from ai_papers.recommender import RecommendationEngine
from ai_papers.llm import _clean_summary_text

print(f"Using repo root: {repo_root}")
```

## 1) Open the Engine

```python
engine = RecommendationEngine(
    data_dir="data",
    categories=["astro-ph.CO", "astro-ph.GA"],
    embedding_model="all-MiniLM-L6-v2",
)
print("Engine ready")
```

## 2) Get Top Recommendations

```python
recs = engine.get_recommendations(limit=5, unrated_only=True)
for i, p in enumerate(recs, start=1):
    print(f"{i}. {p['title']} (score={p.get('score', 0):.3f})")
    print(f"   id={p['arxiv_id']}")
```

## 3) Generate Missing Summaries in Batch

```python
result = engine.generate_missing_summaries(limit=5, include_failed=False)
pprint(result)
```

## 4) Generate One Summary by arXiv ID

```python
if recs:
    one_id = recs[0]['arxiv_id']
    one = engine.generate_summary_for_paper(one_id)
    pprint(one)
else:
    print("No recommendations available yet.")
```

## 5) Text Cleanup Example
The app removes common summary boilerplate before saving text to the database.

```python
raw = "heres a summary of the paper in 3 sentances covering motivation, key findings and limitations: The method improves ranking stability."
print(_clean_summary_text(raw))
```

## 6) Close Resources

```python
engine.close()
print("Engine closed")
```
